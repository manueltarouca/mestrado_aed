```python
descriptive = data.iloc[:,1:5].values
target = data.iloc[:,5].values
print('Descriptive: \n' + str(descriptive))
print('Target: \n' + str(target))
```

```python
from sklearn.preprocessing import Imputer

imputer = Imputer(missing_values='NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(descriptive[:,0:2])
descriptive[:,0:2] = imputer.transform(descriptive[:,0:2])
imputer = imputer.fit(descriptive[:,3:4])
descriptive[:,3:4] = imputer.transform(descriptive[:,3:4])
print('Descriptive: \n' + str(descriptive))
```

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
descriptive[:,2] = labelEncoder.fit_transform(descriptive[:,2])
print('Descriptive: \n' + str(descriptive))
```

```python
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder(categorical_features=[2])
descriptive = oneHotEncoder.fit_transform(descriptive).toarray()
print('Descriptive: \n' + str(descriptive))
```

```python
descriptive.shape
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
descriptive[:,0:8] = scaler.fit_transform(descriptive[:,0:8])
print('Descriptive: \n' + str(descriptive))
```

```python
from sklearn.model_selection import train_test_split
descriptive_training, descriptive_test, target_trining, target_test = train_test_split(descriptive,target,test_size = 0.25, random_state = 0)
print(descriptive_training)
print(descriptive_test)
print(target_trining)
print(target_test)
```
